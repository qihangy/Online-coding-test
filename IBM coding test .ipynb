```python
def get_dic(numberOfAttributes,supportThreshold):
    df = pd.read_csv('/Users/guapofish/Desktop/census.csv',header = None)[:1000]   
    combo = list(itertools.combinations(range(12), numberOfAttributes))
    length_df = len(df)
    output = []
    for item in combo:
        value_count = pd.DataFrame(df[list(item)].agg(','.join, axis=1).value_counts())
        value_count['ratio'] = value_count/length_df
        output.append(list(value_count[value_count['ratio'] >= supportThreshold].index))   
    output_final = list(chain(*output))
    
    return output_final
 
```

```python
get_dic(2,0.6)
```
